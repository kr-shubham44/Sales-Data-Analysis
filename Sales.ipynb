```python
df.dtypes
```

```python
total_sales=df['Sales'].sum()
total_profit=df['Profit'].sum()
total_discount=df['Discount'].sum()
```

## Total Sales, Profit, and Discount


```python
print("Total Sales=",total_sales)
```

```python
print("Total profit=",total_profit)
```

# Region-wise Analysis


## ********* Sales by Region ********* 

```python
region_sales=df.groupby('Region')['Sales'].sum().sort_values(ascending=False)
```

```python
region_sales
```

# VISUALISZATION

```python
plt.figure(figsize=(8,5))
```

```python
sns.barplot(x=region_sales.index, y=region_sales.values, palette='viridis')
```

```python
plt.title("Region-Wise Total Sales")
```

```python
plt.ylabel('Sales')
plt.xlabel('Region')
```

```python
plt.show()
```

# CATEGORY WISE SALES

```python
category_sales=df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
```

```python
category_sales
```

## Visualization

```python
plt.figure(figsize=(9,4))
sns.barplot(x=category_sales.index,y=category_sales.values,palette='coolwarm')
plt.title('CATEGORY WISE SALES')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.show()
```

## Extract month and year

```python
df['Month']=df['Order Date'].dt.month_name()
```

```python
df['Year']=df['Order Date'].dt.year
```

```python
df
```

```python
monthly_sales=df.groupby(['Month','Year'])['Sales'].sum().reset_index()
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(x='Month',y='Sales',hue='Year',data=monthly_sales)
plt.title("Month- Wise Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.yticks(rotation=20)
plt.show()
```

## Profit Discount Analyze

```python
plt.figure(figsize=(9,5))
sns.scatterplot(x='Discount',y='Profit',data=df,alpha=0.6)
plt.title('Profit VS Discount')
plt.show()
```
